```python
import pandas as pd
from deep_translator import GoogleTranslator
from pathlib import Path
import time

data_path = Path(r"C:\Users\azerty\Desktop\Projet2 - Copie\data\processed\movies_fr_final.parquet")
df = pd.read_parquet(data_path)

def traduire(texte):
    if not texte or str(texte).strip().lower() in {"", "none", "nan", "null"}:
        return ""
    try:
        return GoogleTranslator(source='auto', target='fr').translate(str(texte)[:4999])
    except Exception as e:
        print(f"Erreur: {e}")
        return ""

# Trouver les lignes avec erreurs
error_keywords = ["Error 500", "Server Error", "That's an error", "Please try again"]
mask = df["synopsis_fr"].apply(lambda x: any(kw in str(x) for kw in error_keywords) if pd.notna(x) else False)

print(f"{mask.sum()} lignes avec erreurs de traduction")

for i in df[mask].index:
    synopsis_original = df.at[i, "synopsis"]
    df.at[i, "synopsis_fr"] = traduire(str(synopsis_original) if pd.notna(synopsis_original) else "")
    if i % 50 == 0:
        print(f"Corrigé {i}...")
        df.to_parquet(data_path, index=False)
        time.sleep(1)

df.to_parquet(data_path, index=False)
print("Terminé!")
```
