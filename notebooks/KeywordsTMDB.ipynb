```python
import pandas as pd

df = pd.read_parquet("../data/processed/movies_fr_final.parquet")

df.head()
```

```python
import requests
import time

API_KEY = "4b9b9f6d2fe5530cb6fabfbc992e04f6"


def get_keywords(tmdb_id):

    url = f"https://api.themoviedb.org/3/movie/{tmdb_id}/keywords"
    
    params = {"api_key": API_KEY}

    try:
        r = requests.get(url, params=params)
        data = r.json()

        keywords = [k["name"] for k in data.get("keywords", [])]

        time.sleep(0.03)

        return keywords

    except:
        return []
```

```python
from tqdm import tqdm

tqdm.pandas()

df["keywords_list"] = df["id_film"].progress_apply(get_keywords)
```

```python
df.to_parquet(
    "../data/processed/movies_with_keywords.parquet",
    index=False
)
```
